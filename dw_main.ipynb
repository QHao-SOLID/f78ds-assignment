```python
# Loaded variable 'df_enrolment' from URI: data/Primary_School_Enrolments
df_enrolment = pd.read_csv(r'data/Primary_School_Enrolments.csv', skiprows = 4)

# Drop columns: 'Country Name', 'Indicator Name', 'Indicator Code'
df_enrolment = df_enrolment.drop(columns=['Country Name', 'Indicator Name', 'Indicator Code', 'Unnamed: 69'])

# Only keep Malaysia, UAE and United Kingdom
keep_list = ['MYS', 'ARE', 'GBR']
df_enrolment = df_enrolment[df_enrolment['Country Code'].isin(keep_list)]

# Resetting index and drop index column
df_enrolment = df_enrolment.reset_index()
df_enrolment = df_enrolment.drop(columns=['index'])

# Reshaping Dataset
df_enrolment = df_enrolment.melt(
    id_vars=['Country Code'], 
    var_name='Year', 
    value_name='Enrolment Data'
)

# Change column type to int64 for column: 'Year'
df_enrolment = df_enrolment.astype({'Year': 'int64'})

# Simple sort, to satisfy my OCD
df_enrolment = df_enrolment.sort_values(by=['Country Code', 'Year'])
```

```python
df_merged = df_un.merge(df_enrolment, on=['Country Code', 'Year'], how='left')
```

```python
def feature_engineering(df):

    df['Life_Expectancy_Gender_Gap'] = (
        df['Female Life Expectancy at Birth (years)'] - 
        df['Male Life Expectancy at Birth (years)']
    )
    
    df['Child_Survival_Rate'] = 1 - (df['Under-Five Mortality (deaths under age 5 per 1,000 live births)'] / 1000)
    
    df['Migration_Intensity'] = (
        df['Net Number of Migrants (thousands)'] / 
        df['Total Population, as of 1 July (thousands)']
    ) * 100

    # 4. Vital Index (Birth-to-Death Ratio)
    # Measures the natural replenishment of the population. 
    # > 1.0 means more births than deaths.
    df['Vital_Index'] = df['Births (thousands)'] / df['Total Deaths (thousands)']

    # 5. Fertility Replacement Gap
    # Replacement level is generally 2.1. 
    # Negative values indicate a "shrinking" demographic future.
    df['Fertility_Replacement_Gap'] = df['Total Fertility Rate (live births per woman)'] - 2.1

    # 6. Mortality 'Concentration' (Infant vs Under-5)
    # High values mean most child deaths happen in the 1st year (Neonatal/Infant issues)
    # Lower values mean the risk persists throughout early childhood (Nutrition/Environment)
    df['Infant_Mortality_Share_of_U5'] = (
        df['Infant Deaths, under age 1 (thousands)'] / 
        df['Under-Five Deaths, under age 5 (thousands)']
    )

    # 7. Sex Ratio Attrition
    # Compares the ratio at birth vs. the current population ratio.
    # Shows the impact of gender-biased mortality or migration over time.
    df['Sex_Ratio_Attrition'] = (
        df['Population Sex Ratio, as of 1 July (males per 100 females)'] - 
        df['Sex Ratio at Birth (males per 100 female births)']
    )

    return df

# Apply the function
df_merged = feature_engineering(df_merged)
```

```python
def engineer_education_demographics(df):
    """
    Estimates the eligible primary school-age population using the Cohort Survival Method.
    Assumes Primary School Age = 6 to 11 (6-year duration).
    """
    # Ensure data is sorted by country and year for correct shifting
    df = df.sort_values(['Country Code', 'Year'])

    # 1. Calculate the Annual Surviving Cohort
    # We use (Births - Under-Five Deaths) as a proxy for children reaching school age.
    # Note: Using 'Live Births Surviving to Age 1' is also an option if Under-Five is too aggressive.
    df['Annual_Cohort_Surviving_To_5'] = (
        df['Births (thousands)'] - df['Under-Five Deaths, under age 5 (thousands)']
    )

    # 2. Estimate Total Eligible Primary Population (Ages 6-11)
    # Instead of * 6, we sum the surviving cohorts from 6 years ago through 11 years ago.
    # This captures the actual "birth pulses" of the past.
    
    # We create a rolling window of the surviving cohorts, lagged by 6 years.
    df['Est_Eligible_Primary_Population_Thousands'] = (
        df.groupby('Country Code')['Annual_Cohort_Surviving_To_5']
          .transform(lambda x: x.shift(6).rolling(window=6, min_periods=6).sum())
    )

    # 3. Education System "Pressure" Ratio
    # Percentage of the total population that is primary-school age.
    df['Primary_Age_Density_Percent'] = (
        df['Est_Eligible_Primary_Population_Thousands'] / 
        df['Total Population, as of 1 July (thousands)']
    ) * 100

    # 4. Enrolment Gap Analysis
    # Converts 'Est_Eligible' to raw units to match 'Enrolment Data'
    # Result > 0: Over-enrolment (potential late starters/repeaters)
    # Result < 0: Under-enrolment (out-of-school children)
    df['Enrolment_Gap_Units'] = (
        df['Enrolment Data'] - (df['Est_Eligible_Primary_Population_Thousands'] * 1000)
    )

    return df

# Apply the function
df_merged = engineer_education_demographics(df_merged)
```
