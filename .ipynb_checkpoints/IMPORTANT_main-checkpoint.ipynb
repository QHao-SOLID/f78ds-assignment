## Data Cleaning + Merging Process
### UN Population Data Cleaning
- Skipped first 16 rows of metadata during Excel file import
- Dropped 9 unnecessary columns: `Index`, `Variant`, `Region, subregion, country or area *`, `Notes`, `Location code`, `ISO2 Alpha-code`, `SDMX code**`, `Type`, `Parent code`
- Reset index to ensure clean sequential numbering after column removal
- Replaced `...` placeholders with proper NaN values for missing data handling
- Converted 53 demographic columns to float64 data type (population figures, birth/death rates, life expectancy, migration statistics)
- Converted `Year` column to integer type (Int64) for time-series manipulation
- Renamed `ISO3 Alpha-code` to `Country Code` for consistency

### Primary Enrolment Data Cleaning
- Read CSV file skipping first 4 metadata row
- Dropped unnecessary columns: `Country Name`, `Indicator Name`, `Indicator Code`, `Unnamed: 69`
- **Reshaped dataset from wide to tidy format** using melt function:
  - **Original**: separate columns for each year (1960-2024)
  - **New**: `Year` column with years, 'Enrolment Data' column with values
- Converted `Year` column to integer type (Int64)
- Sorted data by `Country Code` and `Year` for logical organization

### Data Merging
- Merged cleaned UN data and enrolment data using left join
- Merge keys: `Country Code` and `Year`
- Left join preserved all UN demographic data even when enrolment data missing

```python
# Data cleaning functions
def clean_un_data(filepath = un_population_filepath):
    # Insert DataFrame from file
    df = pd.read_excel(filepath, skiprows = 16)
    
    # Drop columns: 'Index', 'Variant' and 7 other columns
    df = df.drop(
        columns=[
            'Index', 'Variant',
            'Region, subregion, country or area *',
            'Notes', 'Location code',
            'ISO2 Alpha-code', 'SDMX code**',
            'Type', 'Parent code'
        ]
    )
    
    # Index correction
    df = df.reset_index()
    df = df.drop(columns=['index'])
    
    # Delete unknown strings
    df = df.replace('...', np.nan)
    
    # Change column type to float64 for columns: 'Total Population, as of 1 January (thousands)', 'Total Population, as of 1 July (thousands)' and 52 other columns
    df = df.astype(
        {
            'Total Population, as of 1 January (thousands)': 'float64',
            'Total Population, as of 1 July (thousands)': 'float64',
            'Male Population, as of 1 July (thousands)': 'float64',
            'Female Population, as of 1 July (thousands)': 'float64',
            'Population Density, as of 1 July (persons per square km)': 'float64',
            'Population Sex Ratio, as of 1 July (males per 100 females)': 'float64',
            'Median Age, as of 1 July (years)': 'float64',
            'Natural Change, Births minus Deaths (thousands)': 'float64',
            'Rate of Natural Change (per 1,000 population)': 'float64',
            'Population Change (thousands)': 'float64',
            'Population Growth Rate (percentage)': 'float64',
            'Population Annual Doubling Time (years)': 'float64',
            'Births (thousands)': 'float64',
            'Births by women aged 15 to 19 (thousands)': 'float64',
            'Crude Birth Rate (births per 1,000 population)': 'float64',
            'Total Fertility Rate (live births per woman)': 'float64',
            'Net Reproduction Rate (surviving daughters per woman)': 'float64',
            'Mean Age Childbearing (years)': 'float64',
            'Sex Ratio at Birth (males per 100 female births)': 'float64',
            'Total Deaths (thousands)': 'float64',
            'Male Deaths (thousands)': 'float64',
            'Female Deaths (thousands)': 'float64',
            'Crude Death Rate (deaths per 1,000 population)': 'float64',
            'Life Expectancy at Birth, both sexes (years)': 'float64',
            'Male Life Expectancy at Birth (years)': 'float64',
            'Female Life Expectancy at Birth (years)': 'float64',
            'Life Expectancy at Age 15, both sexes (years)': 'float64',
            'Male Life Expectancy at Age 15 (years)': 'float64',
            'Female Life Expectancy at Age 15 (years)': 'float64',
            'Life Expectancy at Age 65, both sexes (years)': 'float64',
            'Male Life Expectancy at Age 65 (years)': 'float64',
            'Female Life Expectancy at Age 65 (years)': 'float64',
            'Life Expectancy at Age 80, both sexes (years)': 'float64',
            'Male Life Expectancy at Age 80 (years)': 'float64',
            'Female Life Expectancy at Age 80 (years)': 'float64',
            'Infant Deaths, under age 1 (thousands)': 'float64',
            'Infant Mortality Rate (infant deaths per 1,000 live births)': 'float64',
            'Live Births Surviving to Age 1 (thousands)': 'float64',
            'Under-Five Deaths, under age 5 (thousands)': 'float64',
            'Under-Five Mortality (deaths under age 5 per 1,000 live births)': 'float64',
            'Mortality before Age 40, both sexes (deaths under age 40 per 1,000 live births)': 'float64',
            'Male Mortality before Age 40 (deaths under age 40 per 1,000 male live births)': 'float64',
            'Female Mortality before Age 40 (deaths under age 40 per 1,000 female live births)': 'float64',
            'Mortality before Age 60, both sexes (deaths under age 60 per 1,000 live births)': 'float64',
            'Male Mortality before Age 60 (deaths under age 60 per 1,000 male live births)': 'float64',
            'Female Mortality before Age 60 (deaths under age 60 per 1,000 female live births)': 'float64',
            'Mortality between Age 15 and 50, both sexes (deaths under age 50 per 1,000 alive at age 15)': 'float64',
            'Male Mortality between Age 15 and 50 (deaths under age 50 per 1,000 males alive at age 15)': 'float64',
            'Female Mortality between Age 15 and 50 (deaths under age 50 per 1,000 females alive at age 15)': 'float64',
            'Mortality between Age 15 and 60, both sexes (deaths under age 60 per 1,000 alive at age 15)': 'float64',
            'Male Mortality between Age 15 and 60 (deaths under age 60 per 1,000 males alive at age 15)': 'float64',
            'Female Mortality between Age 15 and 60 (deaths under age 60 per 1,000 females alive at age 15)': 'float64',
            'Net Number of Migrants (thousands)': 'float64', 'Net Migration Rate (per 1,000 population)': 'float64'
        }
    )
    
    # Change column type to int64 for column: 'Year'
    df = df.astype({'Year': 'Int64'})
    
    # Rename column 'ISO3 Alpha-code' to 'Country Code'
    df = df.rename(columns={'ISO3 Alpha-code': 'Country Code'})
    
    return df

def clean_enrolment_data(filepath = primary_enrolment_filepath):
    # Insert DataFrame from file
    df = pd.read_csv(filepath, skiprows = 4)
    
    # Drop columns: 'Country Name', 'Indicator Name', 'Indicator Code'
    df = df.drop(
        columns=[
            'Country Name',
            'Indicator Name',
            'Indicator Code',
            'Unnamed: 69'
        ]
    )
    
    # Index correction
    df = df.reset_index()
    df = df.drop(columns=['index'])
    
    # Reshaping Dataset
    df = df.melt(
        id_vars=['Country Code'], 
        var_name='Year', 
        value_name='Enrolment Data'
    )
    
    # Change column type to int64 for column: 'Year'
    df = df.astype({'Year': 'Int64'})
    
    # Simple sort, to satisfy my OCD
    df = df.sort_values(by=['Country Code', 'Year'])
    
    return df

def load_df():
    df_un = clean_un_data()
    df_enrolment = clean_enrolment_data()
    
    return (df_un.merge(df_enrolment, on=['Country Code', 'Year'], how='left'))
```

```python
# Loading cleaned DataFrame
df = load_df()
```

## Feature Engineering
### 1. Life Expectancy Gender Gap
  - Definition: The difference between female and male life expectancy at birth.
  - Purpose: Highlights gender disparities in health outcomes and longevity, which can reflect broader social, economic, and healthcare differences.
  - Calculation: Female Life Expectancy at Birth (years) - Male Life Expectancy at Birth (years).

### 2. Child Survival Rate
  - Definition: The probability that a newborn survives to age five.
  - Purpose: A key indicator of child health, nutrition, and healthcare quality. It is used in estimating the size of future school-age cohorts.
  - Calculation: `1 - (Under‑Five Mortality Rate / 1000)`, where Under‑Five Mortality is expressed per 1,000 live births. The result is rounded to three decimal places.

### 3. Vital Index
  - Definition: The ratio of births to deaths in a given year.
  - Purpose: Provides a quick snapshot of natural population growth pressure. A value greater than 1 indicates more births than deaths (growing population), while less than 1 indicates natural decrease.
  - Calculation: `Births (thousands) / Total Deaths (thousands)`, corrected to three decimal places.

### 4. Annual Cohort Surviving to Age 5
  - Definition: The estimated number of children in a birth year who survive to their fifth birthday.
  - Purpose: This cohort forms the basis for estimating the number of children who will become eligible for primary education in future years.
  - Calculation: Births (thousands) minus Under‑Five Deaths, under age 5 (thousands). The result is in thousands.

### 5. Estimated Eligible Primary Population
  - Definition: The estimated number of children of primary school age in a given year, based on historical births and child survival.
  - Purpose: Serves as a demographic benchmark against which actual enrolment data can be compared to identify gaps or over‑provision.
  - Calculation:
    - For each country, a country‑specific primary school structure is defined:
      - GBR: starts at age 5, duration 6 years ([Source](https://www.gov.uk/schools-admissions/school-starting-age))
      - MYS: starts at age 7, duration 6 years  ([Source](https://www.malaysia.gov.my/en/personas/parents-to-young-children-7-12-years-old/menghantar-anak-ke-sekolah-rendah/primary-education-in-malaysia))
      - ARE: starts at age 6, duration 6 years ([Source](https://u.ae/en/information-and-services/education/school-education-k-12/acceptance-kgs-and-grade-1))
    - The Annual Cohort Surviving to Age 5 is shifted forward by the starting age (e.g., 5 years for GBR) to align births with the year they become eligible.
    - A rolling sum over the duration of primary school (6 years) is then calculated to obtain the total eligible population for each year.
    - Values are stored in thousands and rounded to three decimal places.

### 6. Enrolment Gap
  - Definition: The relative difference between actual primary enrolment and the estimated eligible primary population.
  - Purpose: Quantifies the mismatch between school capacity/demand and the demographic reality. Positive values indicate enrolment exceeds the estimated eligible population (possible over‑enrolment, inclusion of older/younger children, or estimation errors), while negative values indicate under‑enrolment.
  - Calculation:
    - First, the estimated eligible population is converted from thousands to actual numbers (multiply by 1000).
    - Then `(Enrolment Data - Estimated Eligible Population) / Enrolment Data`.
    - The result is a proportion (could be negative or positive), rounded to three decimal places.

```python
# Define country-specific primary structure
primary_structure = {
    'GBR': {'start_age': 5, 'duration': 6},
    'MYS': {'start_age': 7, 'duration': 6},
    'ARE': {'start_age': 6, 'duration': 6}
}
```

```python
# Feature Engineering
def feature_engineering(df):
    df['Life Expectancy Gender Gap'] = (
        df['Female Life Expectancy at Birth (years)'] - 
        df['Male Life Expectancy at Birth (years)']
    )
    
    df['Child Survival Rate'] = 1 - (
        df['Under-Five Mortality (deaths under age 5 per 1,000 live births)'] / 1000
    )

    df['Vital Index'] = df['Births (thousands)'] / df['Total Deaths (thousands)']
    
    df = df.round({
        'Child Survival Rate': 3,
        'Vital Index': 3
    })

    return df

def engineer_education_demographics(df, primary_structure = primary_structure):
    df = df.sort_values(['Country Code', 'Year'])

    df['Annual Cohort Surviving To 5'] = (
        df['Births (thousands)'] - 
        df['Under-Five Deaths, under age 5 (thousands)']
    )
    
    df['Est Eligible Primary Population (thousands)'] = None
    
    for country, params in primary_structure.items():
        
        mask = df['Country Code'] == country
        
        shift_years = params['start_age']
        window_years = params['duration']
        
        df.loc[mask, 'Est Eligible Primary Population (thousands)'] = (
            df.loc[mask]
              .groupby('Country Code')['Annual Cohort Surviving To 5']
              .transform(lambda x: x.shift(shift_years)
                                   .rolling(window=window_years, min_periods=window_years)
                                   .sum())
        )
    
    df['Enrolment Gap'] = (
        (df['Enrolment Data'] -
         (df['Est Eligible Primary Population (thousands)'] * 1000)
        ) / df['Enrolment Data']
    )
    
    df = df.round({
        'Est Eligible Primary Population (thousands)': 3,
        'Enrolment Gap Units': 3
    })
    
    return df

# Apply the function
df = feature_engineering(df)
df = engineer_education_demographics(df)
```

### 1. Description of Population Trends Across All Three Countries
- Malaysia: Follows a steady and consistent growth, but signs of declining is shown after 1980s ([Source](https://georank.org/population/malaysia))
- United Kingdom: Slow, linear growth, due to maturity, lowest growth rate among the other two countries
- UAE: Described as explosive and volatile growth, noticable by surges of growth rates over time, this is due to massive economic migration from oil-discovery in the 1960s ([Source](https://www.indexmundi.com/facts/united-arab-emirates/indicator/SP.URB.TOTL))

```python
countries = ['GBR', 'MYS', 'ARE']

# --- Pivot 1: Total Population ---
df_population = df[df['Country Code'].isin(countries)].pivot(
    index='Year',
    columns='Country Code',
    values='Total Population, as of 1 January (thousands)'
)

# --- Pivot 2: Growth Rate ---
df_growth = df[df['Country Code'].isin(countries)].pivot(
    index='Year',
    columns='Country Code',
    values='Population Growth Rate (percentage)'
)

# --- Create 1 row, 2 columns layout ---
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Left plot: Total Population
df_population.plot(ax=axes[0])
axes[0].set_title("Population Over Time")
axes[0].set_ylabel("Population (thousands)")
axes[0].legend(title="Country")

# Right plot: Growth Rate
df_growth.plot(ax=axes[1])
axes[1].set_title("Population Growth Rate Over Time")
axes[1].set_ylabel("Growth Rate (%)")
axes[1].legend(title="Country")

plt.tight_layout()
plt.show()
```

### Description of Natural Change and Net Migration
- **UK's Mix Population (From Decreasing to Massive Net Migration)**: The UK's shift reflects changes in its economy, policy, and global position. The period of net emigration in the late 1970s and early 1980s coincided with economic restructuring and recession. The shift to massive net migration from the late 1990s onwards is linked to several factors: strong economic growth, labor market demand, EU expansion (which granted free movement rights to new member states), and global conflicts leading to asylum seekers. This transformed the UK from a country of net emigration to a major destination for global migrants.
- **UAE's Low Natural Change and High Net Migration**: This creates a unique demographic structure. The population is heavily skewed towards expatriates, who make up nearly 90% of the total. This leads to a demographic imbalance with a high proportion of working-age males and a much smaller citizen population. The economy is entirely dependent on this continued flow of migrant labor, making its demographic profile a direct reflection of its economic fortunes .
- **Malaysia's Balanced Increasing Trend**: Malaysia represents a classic "developing nation" demographic model. It has successfully transitioned from high birth and death rates to lower ones, but natural increase remains the engine of population growth . The "balanced" nature refers to the fact that its demographic story is an internal one of birth and death rates, rather than being overly swayed by the volatilities of international migration.

```python
countries = ['GBR', 'MYS', 'ARE']

# Create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(18, 12), sharex=True)
axes = axes.flatten()  # Flatten to 1D array for easy looping

for i, country_choice in enumerate(countries):
    
    ax = axes[i]
    df_plot = df[df['Country Code'] == country_choice]
    
    # Stacked bars
    ax.bar(df_plot['Year'],
           df_plot['Natural Change, Births minus Deaths (thousands)'],
           label='Natural Change',
           alpha=0.7)
    
    ax.bar(df_plot['Year'],
           df_plot['Net Number of Migrants (thousands)'],
           bottom=df_plot['Natural Change, Births minus Deaths (thousands)'],
           label='Net Migration',
           alpha=0.7)
    
    # Line for total change
    ax.plot(df_plot['Year'],
            df_plot['Population Change (thousands)'],
            color='black',
            linewidth=2,
            marker='o',
            label='Total Population Change')
    
    ax.set_title(f'Components of Population Change - {country_choice}')
    ax.legend()

# Remove unused subplot (bottom-right)
for j in range(len(countries), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### Estimate vs. Enrolment Data

It should be noted that all three plots follows closely with Natural Change data in the plot above, which could imply the net immigrants are most likely here for workforce / out of eligibility to join primary education.

```python
countries = ['GBR', 'MYS', 'ARE']

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.flatten()  # Make indexing easier

for i, country_choice in enumerate(countries):
    
    df_plot = df[df['Country Code'] == country_choice][
        ['Year',
         'Est Eligible Primary Population (thousands)',
         'Enrolment Data']
    ].copy()
    
    # Convert eligible population from thousands to actual numbers
    df_plot['Est Eligible Primary Population (thousands)'] = (
        df_plot['Est Eligible Primary Population (thousands)'] * 1000
    )
    
    # Prepare data
    df_plot = df_plot.sort_values('Year').set_index('Year')
    
    # Plot
    df_plot.plot.line(
        ax=axes[i],
        title=f"Comparison for {country_choice}"
    )
    
    axes[i].set_ylabel("Population")
    axes[i].legend()

# Remove empty subplot (bottom-right)
fig.delaxes(axes[3])

plt.tight_layout()
plt.show()
```

### Causes of the Enrolment Gap
- **For UAE and United Kingdom**: The significant errors are due to unpredictable migration. Population estimates are often based on projections of natural increase. However, as seen with the UAE's oil boom and the UK's policy shifts, net migration can be sudden, massive, and difficult to forecast. When millions of people move to a country much faster than predicted, the actual school enrollment figures will far exceed the original estimates, creating a large positive "gap."

- **For Malaysia**: The gap for Malaysia narrows over time, with exception of between 2010 to 2020. This likely reflects the successful prediction of steady natural increase. As Malaysia's growth became more stable and predictable, there is less likely of cases with under or over enrolment.

```python
countries = ['GBR', 'MYS', 'ARE']

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.flatten()

for i, country_choice in enumerate(countries):
    
    df_plot = df[df['Country Code'] == country_choice][
        ['Year', 'Enrolment Gap']
    ].copy()
    
    df_plot = df_plot.sort_values('Year').set_index('Year')
    
    # Plot line
    df_plot.plot.line(
        ax=axes[i],
        title=f"Enrolment Gap Comparison - {country_choice}"
    )
    
    axes[i].set_ylim(-0.5, 0.5)

    axes[i].axhline(y=0, linestyle='--')
    
    axes[i].set_ylabel("Enrolment Gap")
    axes[i].legend()

# Remove unused subplot
fig.delaxes(axes[3])

plt.tight_layout()
plt.show()
```
